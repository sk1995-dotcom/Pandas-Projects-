Classification of movies based on Ratings (good,average,excellent)

```python
df.columns
```

```python
def rating(rating):
    if rating>=8.0:
        return "Excelent"
    elif rating >=7.0:
        return "Good"
    else:
        return "Average" 
```

```python
df['Rating_category']=df['Rating'].apply(rating)
df.head()
```

Unique Values From Genre

```python
df.columns
```

```python
df['Genre']
```

```python
list1=[]
for value in df['Genre']:
    list1.append(value.split(','))
list1
```

```python
one_d=[]
for item in list1:
    for item1 in item:
        one_d.append(item1)
one_d
```

```python
unique_list=[]
for item in one_d:
    if item not in unique_list:
        unique_list.append(item)
unique_list
```
